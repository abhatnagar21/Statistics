```python
import numpy as np
import matplotlib.pyplot as plt

# Define multiple parameter sets
params = [
    (1.2, 0.5, 1.2, 0.4, 'r', 'solid', r'$\alpha=1.2, \beta=0.5, \gamma=1.2, p=0.4$'),
    (1.5, 0.8, 1.4, 0.6, 'g', 'dashed', r'$\alpha=1.5, \beta=0.8, \gamma=1.4, p=0.6$'),
    (1.7, 1.2, 1.8, 0.7, 'c', 'dotted', r'$\alpha=1.7, \beta=1.2, \gamma=1.8, p=0.7$'),
    (1.9, 1.8, 2.4, 0.8, 'purple', 'dashdot', r'$\alpha=1.9, \beta=1.8, \gamma=2.4, p=0.8$')
]

# Generate x values
x_values = np.linspace(0.01, 3, 300)  # Avoid zero issues

# Plot the functions for different parameter sets
plt.figure(figsize=(8, 5))
for alpha, beta, gamma, p, color, linestyle, label in params:
    def f(x):
        term = alpha * x + (beta * x**2) / 2
        return (gamma * (alpha + beta * x) * term**(gamma - 1) * 
                (1 - p + p * term*gamma) * np.exp(-term*gamma))

    y_values = f(x_values)
    plt.plot(x_values, y_values, label=label, color=color, linestyle=linestyle)

# Customize plot
plt.xlabel(r'$x$')
plt.ylabel(r'$f(x)$')
plt.title('PDF Plots for Different Parameter Sets')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hazard function definition
def h(x, a, b, g, p):
    z = a*x + b*(x**2)/2.0
    z = np.clip(z, 1e-15, None)  # avoid division by zero
    num = g*(a + b*x)*(z**(g-1)) * (1 - p + p*(z**g))
    den = 1 + p*(z**g)
    return num / den

# Common x-axis
x = np.linspace(0.001, 2.0, 500)

# Line styles & colors
styles = ['-', '--', ':', '-.']
colors = ['red', 'green', 'blue', 'purple']

plt.figure(figsize=(16, 5))

# ------------------- Plot 1 -------------------
params1 = [
    (0.8, 0.5, 1.2, 0.7),
    (1.2, 0.5, 1.4, 0.7),
    (1.5, 0.5, 1.6, 0.7),
    (2.0, 0.5, 1.8, 0.7),
]

plt.subplot(1, 3, 1)
for (a, b, g, p), s, c in zip(params1, styles, colors):
    plt.plot(x, h(x, a, b, g, p), s, color=c, linewidth=2, 
             label=f'={a}, ={g}')
plt.xlabel("x", fontsize=12)
plt.ylabel("HF", fontsize=12)
plt.title("=0.5, p=0.7", fontsize=14)
plt.grid(alpha=0.3)
plt.legend(fontsize=10, loc="upper left")

# ------------------- Plot 2 -------------------
params2 = [
    (1.2, 0.5, 0.1, 0.5),
    (1.2, 1.2, 0.2, 0.5),
    (1.2, 1.5, 0.3, 0.5),
    (1.2, 2.5, 0.4, 0.5),
]

plt.subplot(1, 3, 2)
for (a, b, g, p), s, c in zip(params2, styles, colors):
    plt.plot(x, h(x, a, b, g, p), s, color=c, linewidth=2, 
             label=f'={b}, ={g}')
plt.xlabel("x", fontsize=12)
plt.ylabel("HF", fontsize=12)
plt.title("=1.2, p=0.5", fontsize=14)
plt.grid(alpha=0.3)
plt.legend(fontsize=10, loc="upper right")

# ------------------- Plot 3 -------------------
params3 = [
    (0.2, 0.5, 0.5, 0.8),
    (0.2, 0.5, 0.55, 0.8),
    (0.2, 0.5, 0.6, 0.8),
    (0.2, 0.5, 0.65, 0.8),
]

plt.subplot(1, 3, 3)
for (a, b, g, p), s, c in zip(params3, styles, colors):
    plt.plot(x, h(x, a, b, g, p), s, color=c, linewidth=2, 
             label=f'={g}')
plt.xlabel("x", fontsize=12)
plt.ylabel("HF", fontsize=12)
plt.title("=0.2, =0.5, p=0.8", fontsize=14)
plt.grid(alpha=0.3)
plt.legend(fontsize=10, loc="upper left")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Quadratic function
def f(x, a, b, C):
    return (b/2.0) * x**2 + a*x - C

# Roots function
def roots(a, b, C):
    disc = a**2 + 2*b*C
    if disc < 0:
        return None
    r1 = (-a + np.sqrt(disc)) / b
    r2 = (-a - np.sqrt(disc)) / b
    return r1, r2

# X range
x = np.linspace(-5, 5, 400)

# Parameter sets (, , C)
params = [
    (1.2, 0.5, 2.0),
    (1.5, 0.8, 3.0),
    (0.8, 1.0, 1.5),
    (2.0, 0.3, 4.0)
]

plt.figure(figsize=(14,6))

# ---------------- Plot 1: Quadratic curves ----------------
plt.subplot(1,2,1)
for a, b, C in params:
    y = f(x, a, b, C)
    plt.plot(x, y, label=f"={a}, ={b}, C={C}")
    rts = roots(a, b, C)
    if rts:
        for r in rts:
            plt.plot(r, 0, 'ko')  # mark roots
plt.axhline(0, color="black", linestyle=":")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Quadratic curves for multiple (, , C)")
plt.legend()
plt.grid(alpha=0.3)

# ---------------- Plot 2: Roots vs C ----------------
plt.subplot(1,2,2)
C_values = np.linspace(0.1, 5, 200)
for (a, b, _) in params:
    roots1, roots2 = [], []
    for C in C_values:
        rts = roots(a, b, C)
        if rts:
            roots1.append(rts[0])
            roots2.append(rts[1])
        else:
            roots1.append(np.nan)
            roots2.append(np.nan)
    plt.plot(C_values, roots1, label=f"Root1 (={a}, ={b})")
    plt.plot(C_values, roots2, linestyle="--", label=f"Root2 (={a}, ={b})")

plt.xlabel("C")
plt.ylabel("Roots")
plt.title("Roots as function of C")
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
